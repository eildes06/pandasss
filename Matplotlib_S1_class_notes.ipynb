```python
## Regex
#  text only
#  matches patterns
#  can find matches and substitute  characters
#  can find matching groups
#  produces objects w/ properties and methods that can be interrogated
#  grep  Django   Perl   pd.Series.str   basic str
```

```python
import re
```

```python
pattern = r'\d\d'  # |   &  
target = 'A78L41K'
match = re.search(pattern, target)
```

```python
phone = "2004-959-559 # This is Phone Number"
output = re.sub("\D", "", phone)  #greedy (does many operations)
```

```python
value = "O 1, t 10, o 100. 100000"  #0 -> 999999 
output = re.findall("\d{1,6}", value)
output
```

# Play with width, styles, markers, etc

```python
x = np.arange(0, 10)
x
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(x, x+1, color='red', linewidth=.25)
ax.plot(x, x+2, color='red', linewidth=.5)
ax.plot(x, x+3, color='red', linewidth=.75)
ax.plot(x, x+4, color='red', linewidth=1)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(x, x+1, color='red', lw=.25, linestyle='-')
ax.plot(x, x+2, color='green', lw=.5, ls='-.')
ax.plot(x, x+3, color='k', lw=.75, ls = ':')
ax.plot(x, x+4, color='m', lw=1, ls='--')

#'-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(x, x+1, color='red', marker='+', markersize=20)
ax.plot(x, x+2, color='green',  marker='o', ms=20)
ax.plot(x, x+3, color='k', marker='s', ms=20)
ax.plot(x, x+4, color='m', marker='1', ms=10)
```

```python
x=np.arange(1,10)

fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))

ax[0].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].plot(x,x**2,x,x**3,x,x**4, 'r')

ax[1].set_xlim([1,3])
ax[1].set_ylim([1,50])

ax[0].axvline(x=5,ls="--")
ax[1].axhline(y=15,ls="--")

plt.show()
```
